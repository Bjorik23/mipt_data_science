Найти сумму транзакций за каждый месяц по сферам деятельности, отсортировав по месяцам и по сфере деятельности. — (1 балл)

```python
# SQL-запрос
sql_query = """
SELECT EXTRACT(MONTH FROM t.transaction_date) AS month,
       c.job_industry_category,
       SUM(t.list_price) AS total_transaction_amount
FROM transaction t
JOIN customer c ON t.customer_id = c.customer_id
GROUP BY EXTRACT(MONTH FROM t.transaction_date), c.job_industry_category
ORDER BY EXTRACT(MONTH FROM t.transaction_date), c.job_industry_category;
"""

# Загрузка данных в DataFrame
df = pd.read_sql_query(sql_query, conn)

# Вывод DataFrame
display(df)
```

Найти по всем клиентам сумму всех транзакций (list_price), максимум, минимум и количество транзакций, отсортировав результат по убыванию суммы транзакций и количества клиентов. Выполните двумя способами: используя только group by и используя только оконные функции. Сравните результат. — (2 балла)

```python
# SQL-запрос с GROUP BY
sql_query = """
SELECT 
    c.customer_id,
    SUM(t.list_price) AS total_transactions_amount,
    MAX(t.list_price) AS max_transaction_amount,
    MIN(t.list_price) AS min_transaction_amount,
    COUNT(t.transaction_id) AS transaction_count
FROM 
    customer c
JOIN 
    transaction t ON c.customer_id = t.customer_id
GROUP BY 
    c.customer_id
ORDER BY 
    total_transactions_amount DESC,
    transaction_count DESC;
"""

# Загрузка данных в DataFrame
df = pd.read_sql_query(sql_query, conn)

# Вывод DataFrame
display(df)
```

В этом запросе:
- Мы объединяем таблицы customer и transaction по customer_id.
- Затем мы используем оператор GROUP BY для группировки результатов по customer_id.
- Для каждого клиента вычисляется сумма всех транзакций (total_transactions_amount), максимум (max_transaction_amount) и минимум (min_transaction_amount) стоимости транзакций, а также количество транзакций (transaction_count).
- Результат сортируется сначала по убыванию суммы транзакций (total_transactions_amount) и затем по убыванию количества транзакций (transaction_count).

```python
# SQL-запрос с применением оконных функций
sql_query = """
SELECT 
    customer_id,
    SUM(list_price) OVER (PARTITION BY customer_id) AS total_transactions_amount,
    MAX(list_price) OVER (PARTITION BY customer_id) AS max_transaction_amount,
    MIN(list_price) OVER (PARTITION BY customer_id) AS min_transaction_amount,
    COUNT(transaction_id) OVER (PARTITION BY customer_id) AS transaction_count
FROM 
    transaction
ORDER BY 
    total_transactions_amount DESC,
    transaction_count DESC;
"""

# Загрузка данных в DataFrame
df = pd.read_sql_query(sql_query, conn)

# Вывод DataFrame
display(df)
```

В этом запросе:
- Мы используем оконные функции SUM(), MAX(), MIN() и COUNT() для вычисления суммы, максимума, минимума и количества транзакций для каждого клиента.
- Оператор PARTITION BY используется для разделения результатов на группы по customer_id.
- Результаты не группируются с помощью GROUP BY, но оконные функции применяются ко всем записям в соответствующих окнах (группах), созданных PARTITION BY.
- Результат сортируется так же, как и в первом запросе, по убыванию суммы транзакций (total_transactions_amount) и затем по убыванию количества транзакций (transaction_count).

Оба запроса возвращают аналогичные результаты, но использование оконных функций может быть более гибким, особенно когда требуется выполнить вычисления на уровне группы, но также сохранить информацию о каждой отдельной строке. Найти имена и фамилии клиентов с минимальной/максимальной суммой транзакций за весь период (сумма транзакций не может быть null). Напишите отдельные запросы для минимальной и максимальной суммы. — (2 балла)

```python
# SQL-запрос для минимальной транзакции
sql_query = """
SELECT
    c.first_name,
    c.last_name,
    SUM(t.list_price) AS total_transactions_amount
FROM
    customer c
JOIN
    transaction t ON c.customer_id = t.customer_id
GROUP BY
    c.first_name,
    c.last_name
HAVING
    SUM(t.list_price) = (
        SELECT
            MIN(sum_list_price)
        FROM
            (SELECT
                SUM(list_price) AS sum_list_price
            FROM
                transaction
            GROUP BY
                customer_id) AS subquery
    );
"""

# Загрузка данных в DataFrame
df = pd.read_sql_query(sql_query, conn)

# Вывод DataFrame
display(df)
```

```python
# SQL-запрос для максимальной транзакции
sql_query = """
SELECT
    c.first_name,
    c.last_name,
    SUM(t.list_price) AS total_transactions_amount
FROM
    customer c
JOIN
    transaction t ON c.customer_id = t.customer_id
GROUP BY
    c.first_name,
    c.last_name
HAVING
    SUM(t.list_price) = (
        SELECT
            MAX(sum_list_price)
        FROM
            (SELECT
                SUM(list_price) AS sum_list_price
            FROM
                transaction
            GROUP BY
                customer_id) AS subquery
    );
"""

# Загрузка данных в DataFrame
df = pd.read_sql_query(sql_query, conn)

# Вывод DataFrame
display(df)
```

Вывести только самые первые транзакции клиентов. Решить с помощью оконных функций. — (1 балл)

```python
# SQL-запрос
sql_query = """
SELECT *
FROM (
    SELECT 
        *,
        ROW_NUMBER() OVER (PARTITION BY customer_id ORDER BY transaction_date) AS transaction_rank
    FROM 
        transaction
) AS ranked_transactions
WHERE 
    transaction_rank = 1;
"""

# Загрузка данных в DataFrame
df = pd.read_sql_query(sql_query, conn)

# Вывод DataFrame
display(df)
```

Вывести имена, фамилии и профессии клиентов, между транзакциями которых был максимальный интервал (интервал вычисляется в днях) — (2 балла).

```python
# SQL-запрос
sql_query = """
WITH transaction_gaps AS (
    SELECT
        customer_id,
        transaction_date,
        LAG(transaction_date) OVER (PARTITION BY customer_id ORDER BY transaction_date) AS prev_transaction_date,
        transaction_date - LAG(transaction_date) OVER (PARTITION BY customer_id ORDER BY transaction_date) AS gap
    FROM
        transaction
),
max_gap AS (
    SELECT
        MAX(gap) AS max_gap
    FROM
        transaction_gaps
)
SELECT
    c.first_name,
    c.last_name,
    c.job_title,
    mg.max_gap
FROM
    customer c
JOIN
    transaction_gaps tg ON c.customer_id = tg.customer_id
CROSS JOIN
    max_gap mg
WHERE
    tg.gap = mg.max_gap
ORDER BY
    c.first_name,
    c.last_name;
"""

# Загрузка данных в DataFrame
df = pd.read_sql_query(sql_query, conn)

# Вывод DataFrame
display(df)
```
